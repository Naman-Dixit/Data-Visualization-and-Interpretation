```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic=sns.load_dataset("titanic")
```

```python
# 1. Survivors vs Non-survivors
plt.figure(figsize=(6,4))
sns.countplot(data=titanic, x='survived', palette='Set2')
plt.title("Survivors vs Non-Survivors")
plt.xlabel("Survival Status (0 = Died, 1 = Survived)")
plt.ylabel("Count")
plt.show()
# Observation: More people did not survive compared to those who did.
```

```python
# 2. Survival rate by gender
plt.figure(figsize=(6,4))
sns.barplot(data=titanic, x='sex', y='survived', palette='pastel')
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
# Observation: Females had a much higher survival rate compared to males.
```

```python
# 3. Heatmap of survival rate by passenger class
survival_pclass = titanic.pivot_table(values='survived', index='class', aggfunc='mean')
plt.figure(figsize=(6,4))
sns.heatmap(survival_pclass, annot=True, cmap='Blues', cbar=False)
plt.title("Survival Rate by Passenger Class")
plt.show()
# Observation: Higher-class passengers (1st class) had higher survival rates compared to 2nd and 3rd class.
```

```python
# 4. Survival counts across classes
plt.figure(figsize=(6,4))
sns.countplot(data=titanic, x='class', hue='survived', palette='Set1')
plt.title("Survival Count by Passenger Class")
plt.ylabel("Count")
plt.show()
# Observation: Majority of deaths occurred in 3rd class, while more 1st-class passengers survived.
```

```python
# 5. Age distribution of survivors vs non-survivors
plt.figure(figsize=(8,5))
sns.violinplot(data=titanic, x='survived', y='age', palette='muted', split=True)
plt.title("Age Distribution of Survivors vs Non-Survivors")
plt.xlabel("Survival Status (0 = Died, 1 = Survived)")
plt.ylabel("Age")
plt.show()
# Observation: Younger passengers had slightly better survival chances. Many children survived.
```

```python
# 6. Boxplot for age distribution by survival status
plt.figure(figsize=(6,4))
sns.boxplot(data=titanic, x='survived', y='age', palette='Set3')
plt.title("Age Distribution by Survival Status")
plt.xlabel("Survival Status (0 = Died, 1 = Survived)")
plt.ylabel("Age")
plt.show()
# Observation: The median age of survivors was lower compared to non-survivors. Outliers exist in both groups (elderly passengers).
```
