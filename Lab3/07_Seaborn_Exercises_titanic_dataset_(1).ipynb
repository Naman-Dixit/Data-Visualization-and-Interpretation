```python
# Drop missing values for clean plotting
titanic = titanic.dropna(subset=["age", "fare"])

# Jointplot with regression and correlation
sns.jointplot(
    x="fare",
    y="age",
    data=titanic,
    kind="scatter",
    height=6,
    marginal_kws=dict(bins=25, fill=True, color="lightblue")
)

plt.suptitle("Joint Distribution of Fare vs Age (Titanic Dataset)", y=1.02)
plt.show()
```

```python
 #Compute correlation matrix (only numeric columns)
corr = titanic.corr(numeric_only=True)

# Heatmap of correlations
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=False, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap (Titanic Dataset)")
plt.show()
```

```python
# Create subplots for male and female
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Filter data by sex and remove NaN ages
male_ages = titanic[titanic['sex'] == 'male']['age'].dropna()
female_ages = titanic[titanic['sex'] == 'female']['age'].dropna()

# Create histogram for males
ax1.hist(male_ages, bins=range(0, 85, 5), alpha=0.7, color='steelblue', edgecolor='black', linewidth=0.5)
ax1.set_title('sex = male', fontsize=12)
ax1.set_xlabel('age')
ax1.set_ylabel('count')
ax1.set_xlim(0, 80)
ax1.set_ylim(0, 120)
ax1.grid(True, alpha=0.3)
ax1.set_xticks(range(0, 90, 10))
ax1.set_yticks(range(0, 140, 20))

# Create histogram for females
ax2.hist(female_ages, bins=range(0, 85, 5), alpha=0.7, color='steelblue', edgecolor='black', linewidth=0.5)
ax2.set_title('sex = female', fontsize=12)
ax2.set_xlabel('age')
ax2.set_xlim(0, 80)
ax2.set_ylim(0, 120)
ax2.grid(True, alpha=0.3)
ax2.set_xticks(range(0, 90, 10))
ax2.set_yticks(range(0, 140, 20))

# Adjust layout and display
plt.tight_layout()
plt.show()
```
