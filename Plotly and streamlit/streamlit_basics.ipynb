## Create the streamlit app

### Subtask:
Write the Python code for your Streamlit dashboard. This code will define the layout, components, and logic of your dashboard.


**Reasoning**:
Write the Python code for the Streamlit dashboard, including importing streamlit, adding a title and header, adding interactive elements, and displaying output based on the interactive elements.



```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np

st.title("My Simple Streamlit Dashboard with Plots")
st.write("This is a simple dashboard to demonstrate basic Streamlit features and plotting.")

# Add a slider
age = st.slider("Select your age", 0, 100, 25)
st.write("Your age is:", age)

# Add a selectbox
option = st.selectbox(
    "Select your favorite color",
    ("Red", "Blue", "Green", "Yellow")
)
st.write("Your favorite color is:", option)

# Add a text input
name = st.text_input("Enter your name", "Anonymous")
st.write("Hello,", name)

# Add a section for plots
st.header("Basic Plots")

# Generate some sample data for plotting
chart_data = pd.DataFrame(
    np.random.randn(20, 3),
    columns=['a', 'b', 'c']
)

# Add a line chart
st.subheader("Line Chart")
st.line_chart(chart_data)

# Add a bar chart
st.subheader("Bar Chart")
st.bar_chart(chart_data)
```
