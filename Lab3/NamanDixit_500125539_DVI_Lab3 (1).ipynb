```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set Seaborn style
sns.set_theme(style="darkgrid")
```

```python
# Section 1: Visualization over Built-in Datasets
# Load datasets
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
iris = sns.load_dataset("iris")

# ----- Tips Dataset -----
print("Tips Dataset:\n", tips.head())

# Histogram of total bill
plt.figure(figsize=(8,5))
sns.histplot(tips['total_bill'], kde=True, color="skyblue")
plt.title("Histogram of Total Bill")
plt.show()

# Scatter plot of total_bill vs tip
plt.figure(figsize=(8,5))
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips, palette="Set2")
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()

# ----- Flights Dataset -----
print("\nFlights Dataset:\n", flights.head())

# Pivot for heatmap
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Heatmap of passengers
plt.figure(figsize=(12,6))
sns.heatmap(flights_pivot, cmap="Blues", annot=True, fmt="d")
plt.title("Heatmap of Passengers (Flights Dataset)")
plt.show()

# ----- Iris Dataset -----
print("\nIris Dataset:\n", iris.head())

# Pairplot
sns.pairplot(iris, hue="species", palette="husl")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
# Section 2: Practice with Different Plot Categories
# 1. Distribution Plots
plt.figure(figsize=(8,5))
sns.histplot(iris['sepal_length'], kde=True, color="green")
plt.title("Distribution of Sepal Length (Iris)")
plt.show()

plt.figure(figsize=(8,5))
sns.kdeplot(x=iris['petal_width'], fill=True, color="purple")
plt.title("KDE of Petal Width (Iris)")
plt.show()

# 2. Categorical Plots
plt.figure(figsize=(8,5))
sns.barplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("Average Total Bill by Day (Tips)")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x="sex", data=tips, palette="Set1")
plt.title("Count of Customers by Sex (Tips)")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x="day", y="tip", hue="sex", data=tips, palette="coolwarm")
plt.title("Boxplot of Tips by Day and Gender")
plt.show()

plt.figure(figsize=(8,5))
sns.violinplot(x="species", y="petal_length", data=iris, palette="muted")
plt.title("Violin Plot of Petal Length by Species (Iris)")
plt.show()

plt.figure(figsize=(8,5))
sns.swarmplot(x="species", y="petal_width", data=iris, palette="Set2")
plt.title("Swarmplot of Petal Width by Species (Iris)")
plt.show()

# 3. Matrix Plots (Flights Dataset)
plt.figure(figsize=(12,6))
sns.clustermap(flights_pivot, cmap="coolwarm", standard_scale=1)
plt.suptitle("Clustermap of Flights Passengers", y=1.02)
plt.show()

# 4. Regression Plots (Tips Dataset)
plt.figure(figsize=(8,5))
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"color":"blue"}, line_kws={"color":"red"})
plt.title("Regression Plot: Total Bill vs Tip")
plt.show()

sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips, aspect=1.2, height=6, palette="Set2")
plt.suptitle("LMPlot: Total Bill vs Tip by Sex", y=1.02)
plt.show()
```

```python
# Section 3: Grid, Style, and Color Customization
# 1. PairGrid, FacetGrid, and JointGrid

# PairGrid with Iris dataset
g = sns.PairGrid(iris, hue="species", palette="Set2")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
g.add_legend()
g.fig.suptitle("PairGrid of Iris Dataset", y=1.02)
plt.show()

# FacetGrid with Tips dataset (scatterplots for smokers/non-smokers by time)
g = sns.FacetGrid(tips, col="time", row="smoker", margin_titles=True, palette="Set1")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip", hue="sex")
g.add_legend()
g.fig.suptitle("FacetGrid of Tips Dataset by Time and Smoker", y=1.02)
plt.show()

# JointGrid with Tips dataset
g = sns.JointGrid(data=tips, x="total_bill", y="tip", space=0, height=6, ratio=50)
g.plot(sns.scatterplot, sns.histplot)
g.plot_joint(sns.kdeplot, fill=True, thresh=0.1)
g.set_axis_labels("Total Bill", "Tip")
plt.suptitle("JointGrid: Total Bill vs Tip (Tips Dataset)", y=1.02)
plt.show()

# 2. Experiment with Seaborn styles
for style in ["darkgrid", "whitegrid", "ticks"]:
    sns.set_style(style)
    plt.figure(figsize=(8,5))
    sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris, palette="husl")
    plt.title(f"Iris Scatterplot with style = {style}")
    plt.xlabel("Sepal Length")
    plt.ylabel("Sepal Width")
    plt.legend(title="Species")
    plt.show()

# Reset style
sns.set_theme(style="darkgrid")

# 3. Apply different color palettes
for palette in ["Set1", "viridis", "coolwarm", "magma"]:
    plt.figure(figsize=(8,5))
    sns.boxplot(x="day", y="total_bill", hue="sex", data=tips, palette=palette)
    plt.title(f"Boxplot of Total Bill by Day (Palette = {palette})")
    plt.xlabel("Day of Week")
    plt.ylabel("Total Bill")
    plt.legend(title="Sex")
    plt.show()

# 4. Add Titles, Axis Labels, Legends (Example enhancements)

# Histogram with labels
plt.figure(figsize=(8,5))
sns.histplot(tips["total_bill"], kde=True, color="orange")
plt.title("Distribution of Total Bill (Tips Dataset)", fontsize=14)
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.show()

# Heatmap with labels
plt.figure(figsize=(12,6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", cbar_kws={'label': 'Number of Passengers'})
plt.title("Flights Heatmap (Passengers per Month and Year)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()

# Regression plot with legend
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips, height=6, aspect=1.2, palette="Set2")
plt.title("Regression: Total Bill vs Tip by Sex", fontsize=16)
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```
