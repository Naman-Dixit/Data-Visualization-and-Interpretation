```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
tips = sns.load_dataset("tips")

# Preview data
print(tips.head())
```

```python
plt.figure(figsize=(12,5))

# Total bill distribution
plt.subplot(1,2,1)
sns.histplot(tips['total_bill'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Total Bill")

# Tip distribution
plt.subplot(1,2,2)
sns.histplot(tips['tip'], bins=30, kde=True, color='orange')
plt.title("Distribution of Tip")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='time', y='tip', data=tips, palette="Set2")
plt.title("Tip Amount by Time of Day (Lunch vs Dinner)")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='day', y='tip', data=tips, palette="pastel")
plt.title("Tip Amount by Day of Week")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x="sex", y="tip", data=tips, palette="coolwarm")
plt.title("Tipping Behavior: Male vs Female")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x="day", y="tip", hue="time", data=tips, jitter=True, palette="Set1", dodge=True)
plt.title("Tips by Day & Time (Highlighting Extreme Tips)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, palette="tab10", s=80)
plt.title("Total Bill vs Tip by Day")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="day", y="tip", data=tips, palette="muted")
plt.title("Distribution & Density of Tips Across Days")
plt.show()
```

```python
# Create contingency table
day_time = pd.crosstab(tips['day'], tips['time'])

plt.figure(figsize=(6,5))
sns.heatmap(day_time, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Frequency Heatmap of Days vs Time")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x="day", y="tip", data=tips, estimator=lambda x: sum(x)/len(x), palette="Set3")
plt.title("Average Tip Amount by Day of Week")
plt.show()
```

**Report on TIPS Dataset Visualization and Observations**

**1. Distribution of Total Bill and Tip**

The total bill values are mostly concentrated between $10 and $30, with a few extreme values above $50, indicating the presence of outliers.

The tip values are mainly between $2  and  $5, with a few very high tips above $10, which are unusual and could be linked to higher bills.

Both distributions are right-skewed, suggesting that small to moderate bills and tips are more common than large ones.

**2. Variation of Tip Amount by Time**

(i) Lunch vs Dinner

The median tip during Dinner is noticeably higher than during Lunch.

Dinner service also shows more variability in tips, with some extreme high values.

Lunch tips are generally lower and more consistent.

(ii) Day of the Week

Sunday and Saturday show higher and more varied tips compared to weekdays.

Thursday has the lowest spread of tips, indicating less variation.

Overall, tips tend to be higher on weekends when more people dine out.

**3. Tipping Behavior: Male vs Female**

Both male and female customers give a similar range of tips, but males tend to tip slightly higher on average.

The spread of tips among males is wider, indicating that extreme tip values are more common among male customers.

Females generally give tips in a narrower and more consistent range.

**4. Concentration of Highest and Lowest Tips**

Highest tips are clustered around Saturday dinners and Sunday dinners, aligning with higher weekend dining activity.

Lowest tips are more scattered but occur more frequently during weekday lunches.

This suggests weekend evenings encourage more generous tipping behavior.

**5. Relationship Between Total Bill and Tip (Scatter Plot)**

There is a positive correlation: as the total bill increases, the tip also increases.

However, the increase is not perfectly linearsome large bills have relatively smaller tips, indicating variations in tipping generosity.

Each day shows a similar pattern, though Saturday and Sunday scatter points include the largest bills and tips.

**6. Distribution of Tips Across Days (Violin Plot)**

Saturday and Sunday have wider distributions, showing greater variability in tipping behavior.

Thursday and Friday show narrower, more consistent distributions.

The density peaks suggest that $2$4 tips are the most common across all days.

**7. Heatmap of Days vs Time**

Dinner dominates the dataset across all days, especially on Saturday and Sunday.

Lunch data is less frequent and mainly concentrated on Thursday and Friday.

This explains why dinner tips show more variationit is the most common dining occasion.

**8. Generosity of Tips by Day**

Saturday records the highest average tip amount, followed closely by Sunday.

Thursday has the lowest average tip amount, with Friday being slightly higher.

This indicates that customers tend to be more generous with tips on weekends compared to weekdays.

**Conclusion**

The TIPS dataset shows clear behavioral patterns in tipping.

Weekend dinners are associated with higher and more generous tips, while weekday lunches often show smaller, consistent tips.

Males tend to give slightly higher and more variable tips than females.

Outliers exist in both total bills and tips, mostly tied to larger weekend dining bills.

Overall, tipping is positively correlated with bill size, but customer generosity also depends on time of day, day of week, and gender.
