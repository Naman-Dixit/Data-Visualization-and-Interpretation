## Create visualizations

### Subtask:
Create some visualizations using the dataset.


**Reasoning**:
Create the requested visualizations using seaborn.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Total Bill vs Tip')

plt.subplot(1, 3, 2)
sns.boxplot(x='day', y='tip', data=df)
plt.title('Tip by Day')

plt.subplot(1, 3, 3)
sns.histplot(data=df, x='total_bill')
plt.title('Distribution of Total Bill')

plt.tight_layout()
plt.show()
```

## Build the dashboard

### Subtask:
Combine the visualizations into a dashboard.


**Reasoning**:
Combine the previously created visualizations into a single figure with subplots to create a dashboard layout.



```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.scatterplot(x='total_bill', y='tip', data=df, ax=axes[0])
axes[0].set_title('Total Bill vs Tip')

sns.boxplot(x='day', y='tip', data=df, ax=axes[1])
axes[1].set_title('Tip by Day')

sns.histplot(data=df, x='total_bill', ax=axes[2])
axes[2].set_title('Distribution of Total Bill')

plt.tight_layout()
plt.show()
```

## Create visualizations with streamlit

```python
def main():
    st.set_page_config(page_title="Tips Dashboard", layout="wide")
    st.title("Tips Dataset Dashboard")
    st.write("This dashboard visualizes the 'tips' dataset from the seaborn library.")

    st.markdown("## Raw Data")
    st.dataframe(df)

    st.markdown("## Visualizations")

    # Scatter plot
    fig1, ax1 = plt.subplots(figsize=(8, 5))
    sns.scatterplot(x='total_bill', y='tip', data=df, ax=ax1)
    ax1.set_title('Total Bill vs Tip')
    st.pyplot(fig1)

    # Box plot
    fig2, ax2 = plt.subplots(figsize=(8, 5))
    sns.boxplot(x='day', y='tip', data=df, ax=ax2)
    ax2.set_title('Tip by Day')
    st.pyplot(fig2)

    # Histogram
    fig3, ax3 = plt.subplots(figsize=(8, 5))
    sns.histplot(data=df, x='total_bill', ax=ax3)
    ax3.set_title('Distribution of Total Bill')
    st.pyplot(fig3)


if __name__ == "__main__":
    main()
```

## Add interactive widgets

### Subtask:
Include interactive elements like selectboxes or sliders to filter the data or change plot parameters.


**Reasoning**:
Add a sidebar with selectboxes for filtering 'day' and 'sex', filter the dataframe based on selections, and update the visualizations to use the filtered data.



```python
def main():
    st.set_page_config(page_title="Tips Dashboard", layout="wide")
    st.title("Tips Dataset Dashboard")
    st.write("This dashboard visualizes the 'tips' dataset from the seaborn library.")

    st.markdown("## Raw Data")
    st.dataframe(df)

    st.markdown("## Visualizations")

    # Sidebar for filtering
    st.sidebar.header("Filter Data")
    day_filter = st.sidebar.selectbox("Select Day", options=['All'] + list(df['day'].unique()))
    sex_filter = st.sidebar.selectbox("Select Sex", options=['All'] + list(df['sex'].unique()))

    filtered_df = df.copy()
    if day_filter != 'All':
        filtered_df = filtered_df[filtered_df['day'] == day_filter]
    if sex_filter != 'All':
        filtered_df = filtered_df[filtered_df['sex'] == sex_filter]

    # Scatter plot
    fig1, ax1 = plt.subplots(figsize=(8, 5))
    sns.scatterplot(x='total_bill', y='tip', data=filtered_df, ax=ax1)
    ax1.set_title('Total Bill vs Tip')
    st.pyplot(fig1)

    # Box plot
    fig2, ax2 = plt.subplots(figsize=(8, 5))
    sns.boxplot(x='day', y='tip', data=filtered_df, ax=ax2)
    ax2.set_title('Tip by Day')
    st.pyplot(fig2)

    # Histogram
    fig3, ax3 = plt.subplots(figsize=(8, 5))
    sns.histplot(data=filtered_df, x='total_bill', ax=ax3)
    ax3.set_title('Distribution of Total Bill')
    st.pyplot(fig3)


if __name__ == "__main__":
    main()
```

## Organize layout

### Subtask:
Arrange the elements in the dashboard for better presentation using Streamlit's layout features.


**Reasoning**:
Arrange the elements in the dashboard for better presentation using Streamlit's layout features.



```python
def main():
    st.set_page_config(page_title="Tips Dashboard", layout="wide")
    st.title("Tips Dataset Dashboard")
    st.write("This dashboard visualizes the 'tips' dataset from the seaborn library.")

    # Sidebar for filtering
    st.sidebar.header("Filter Data")
    day_filter = st.sidebar.selectbox("Select Day", options=['All'] + list(df['day'].unique()))
    sex_filter = st.sidebar.selectbox("Select Sex", options=['All'] + list(df['sex'].unique()))

    filtered_df = df.copy()
    if day_filter != 'All':
        filtered_df = filtered_df[filtered_df['day'] == day_filter]
    if sex_filter != 'All':
        filtered_df = filtered_df[filtered_df['sex'] == sex_filter]

    # Raw Data Section (Collapsible)
    with st.expander("View Raw Data"):
        st.dataframe(filtered_df)

    st.markdown("## Visualizations")

    # Arrange visualizations in columns
    col1, col2, col3 = st.columns(3)

    with col1:
        # Scatter plot
        fig1, ax1 = plt.subplots(figsize=(6, 4))
        sns.scatterplot(x='total_bill', y='tip', data=filtered_df, ax=ax1)
        ax1.set_title('Total Bill vs Tip')
        st.pyplot(fig1)
        plt.close(fig1) # Close the figure to free memory

    with col2:
        # Box plot
        fig2, ax2 = plt.subplots(figsize=(6, 4))
        sns.boxplot(x='day', y='tip', data=filtered_df, ax=ax2)
        ax2.set_title('Tip by Day')
        st.pyplot(fig2)
        plt.close(fig2) # Close the figure to free memory

    with col3:
        # Histogram
        fig3, ax3 = plt.subplots(figsize=(6, 4))
        sns.histplot(data=filtered_df, x='total_bill', ax=ax3)
        ax3.set_title('Distribution of Total Bill')
        st.pyplot(fig3)
        plt.close(fig3) # Close the figure to free memory


if __name__ == "__main__":
    main()
```
